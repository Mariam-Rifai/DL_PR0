```python
import torch
import torch.nn as nn
import torch.optim as optim

import torchvision
import torchvision.transforms as transforms

from torch.utils.data import DataLoader

import matplotlib.pyplot as plt

# =========================
# 2) Data Preprocessing
# =========================

transform = transforms.Compose([

    transforms.ToTensor(),

    transforms.Normalize((0.5,), (0.5,))
])

# =========================
# 3) Load Dataset
# =========================

train_dataset = torchvision.datasets.MNIST(
    root='./data',
    train=True,
    transform=transform,
    download=True
)

test_dataset = torchvision.datasets.MNIST(
    root='./data',
    train=False,
    transform=transform,
    download=True
)

# =========================
# 4) DataLoader
# =========================

train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False
)

print("Train Data:", len(train_dataset))
print("Test Data:", len(test_dataset))

# =========================
# 5) Show Sample Images
# =========================

images, labels = next(iter(train_loader))

plt.figure(figsize=(10,5))

for i in range(6):

    plt.subplot(2,3,i+1)

    plt.imshow(images[i].squeeze(), cmap='gray')

    plt.title(f"Label: {labels[i].item()}")

    plt.axis("off")

plt.show()

# ==========================================
# 6) Experiment 1
# CNN + ReLU
# ==========================================

model1 = nn.Sequential(

    # -------- Convolution Block 1 --------

    nn.Conv2d(1,16,3,padding=1),

    nn.ReLU(),

    nn.MaxPool2d(2),

    # -------- Convolution Block 2 --------

    nn.Conv2d(16,32,3,padding=1),

    nn.ReLU(),

    nn.MaxPool2d(2),

    # -------- Fully Connected --------

    nn.Flatten(),

    nn.Linear(32*7*7,128),

    nn.ReLU(),

    nn.Dropout(0.5),

    nn.Linear(128,10)
)

print(model1)

criterion1 = nn.CrossEntropyLoss()

optimizer1 = optim.Adam(
    model1.parameters(),
    lr=0.001
)

train_losses1 = []
accuracies1 = []

val_losses1 = []
val_accuracies1 = []

epochs = 5

print("\nTraining Experiment 1...\n")

for epoch in range(epochs):

    model1.train()

    running_loss = 0

    correct = 0
    total = 0

    # -------- Training --------

    for images, labels in train_loader:

        outputs = model1(images)

        loss = criterion1(outputs, labels)

        optimizer1.zero_grad()

        loss.backward()

        optimizer1.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs,1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

    accuracy = correct / total

    train_losses1.append(running_loss)

    accuracies1.append(accuracy)

    # -------- Validation --------

    model1.eval()

    val_loss = 0

    val_correct = 0
    val_total = 0

    with torch.no_grad():

        for images, labels in test_loader:

            outputs = model1(images)

            loss = criterion1(outputs, labels)

            val_loss += loss.item()

            _, predicted = torch.max(outputs,1)

            val_total += labels.size(0)

            val_correct += (predicted == labels).sum().item()

    val_accuracy = val_correct / val_total

    val_losses1.append(val_loss)

    val_accuracies1.append(val_accuracy)

    print(
        f"Epoch {epoch+1} | "
        f"Train Loss: {running_loss:.4f} | "
        f"Train Accuracy: {accuracy:.4f} | "
        f"Val Loss: {val_loss:.4f} | "
        f"Val Accuracy: {val_accuracy:.4f}"
    )

# -------- Final Testing --------

model1.eval()

correct = 0
total = 0

test_loss1 = 0

with torch.no_grad():

    for images, labels in test_loader:

        outputs = model1(images)

        loss = criterion1(outputs, labels)

        test_loss1 += loss.item()
        _, predicted = torch.max(outputs,1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

test_acc1 = correct / total

print("\nExperiment 1 Test Accuracy:", test_acc1)

# ==========================================
# 7) Experiment 2
# CNN + Tanh
# ==========================================

model2 = nn.Sequential(

    # -------- Convolution Block 1 --------

    nn.Conv2d(1,16,3,padding=1),

    nn.Tanh(),

    nn.MaxPool2d(2),

    # -------- Convolution Block 2 --------

    nn.Conv2d(16,32,3,padding=1),

    nn.Tanh(),

    nn.MaxPool2d(2),

    # -------- Fully Connected --------

    nn.Flatten(),

    nn.Linear(32*7*7,128),

    nn.Tanh(),

    nn.Dropout(0.5),

    nn.Linear(128,10)
)

print(model2)

criterion2 = nn.CrossEntropyLoss()

optimizer2 = optim.Adam(
    model2.parameters(),
    lr=0.001
)

train_losses2 = []
accuracies2 = []

val_losses2 = []
val_accuracies2 = []

epochs = 5

print("\nTraining Experiment 2...\n")

for epoch in range(epochs):

    model2.train()

    running_loss = 0

    correct = 0
    total = 0

    # -------- Training --------

    for images, labels in train_loader:

        outputs = model2(images)

        loss = criterion2(outputs, labels)

        optimizer2.zero_grad()

        loss.backward()

        optimizer2.step()

        running_loss += loss.item()

        _, predicted = torch.max(outputs,1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

    accuracy = correct / total

    train_losses2.append(running_loss)

    accuracies2.append(accuracy)

    # -------- Validation --------

    model2.eval()

    val_loss = 0

    val_correct = 0
    val_total = 0

    with torch.no_grad():

        for images, labels in test_loader:

            outputs = model2(images)

            loss = criterion2(outputs, labels)

            val_loss += loss.item()

            _, predicted = torch.max(outputs,1)

            val_total += labels.size(0)

            val_correct += (predicted == labels).sum().item()

    val_accuracy = val_correct / val_total

    val_losses2.append(val_loss)

    val_accuracies2.append(val_accuracy)

    print(
        f"Epoch {epoch+1} | "
        f"Train Loss: {running_loss:.4f} | "
        f"Train Accuracy: {accuracy:.4f} | "
        f"Val Loss: {val_loss:.4f} | "
        f"Val Accuracy: {val_accuracy:.4f}"
    )

# -------- Final Testing --------

model2.eval()

correct = 0
total = 0

test_loss2 = 0

with torch.no_grad():

    for images, labels in test_loader:

        outputs = model2(images)

        loss = criterion2(outputs, labels)

        test_loss2 += loss.item()

        _, predicted = torch.max(outputs,1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

test_acc2 = correct / total

print("\nExperiment 2 Test Accuracy:", test_acc2)

# ==========================================
# 8) Loss Curves
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(train_losses1, label="Experiment 1 Loss")

plt.plot(val_losses1, label="Experiment 1 Val Loss")

plt.plot(train_losses2, label="Experiment 2 Loss")

plt.plot(val_losses2, label="Experiment 2 Val Loss")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.title("Loss Curves")

plt.legend()

plt.show()

# ==========================================
# 9) Accuracy Curves
# ==========================================

plt.figure(figsize=(10,5))

plt.plot(accuracies1, label="Experiment 1 Accuracy")

plt.plot(val_accuracies1, label="Experiment 1 Val Accuracy")

plt.plot(accuracies2, label="Experiment 2 Accuracy")

plt.plot(val_accuracies2, label="Experiment 2 Val Accuracy")

plt.xlabel("Epoch")

plt.ylabel("Accuracy")

plt.title("Accuracy Curves")

plt.legend()

plt.show()

# ==========================================
# 10) Final Results
# ==========================================

print("\n========== RESULTS ==========")

print(f"Experiment 1 Accuracy: {test_acc1:.4f}")

print(f"Experiment 2 Accuracy: {test_acc2:.4f}")

if test_acc1 > test_acc2:
  print("\nExperiment 1 performed better.")

else:

    print("\nExperiment 2 performed better.")
```
